2. Remove the columns that are not needed

```python
df.drop(columns="AreaName") # drop single value
```

```python
df.drop(index=[0,2,4,6,8]) # rows are removed when index number are given
```

```python
cols_to_drop = ["Type", "Coverage", "AREA", "REG", "DEV"]
df.drop(columns=cols_to_drop, inplace=True) # it will apply the results to dataframe itself
```

3. Rename the columns according to yourself

```python
df.columns
```

```python
df.rename({
    "OdName":"Country",
    "AreaName":"Continent",
    "RegName":"Region",
    "DevName":"Status"
}, axis=1, inplace=True) 
df.head() 
```

4. Set the index of the dataframe if it is needed

```python
df.set_index("Country", inplace=True)
df.head()
```

6. Add a new column to the dataframe

```python
years = list(range(1980,2014))
df["total"] = df[years].sum(axis=1)
```
