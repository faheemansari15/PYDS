```python
import requests
from bs4 import BeautifulSoup
```

```python
def load_soup(url):
    page = requests.get(url)
    if page.status_code == 200:
        return BeautifulSoup(page.text)
```

```python
url = "https://www.flipkart.com/search?q=mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
soup = load_soup(url)
if soup is not None:
    print("Data loaded") 
else:
    print("Data could not be loaded")    
```

```python
target = soup.find("div", attrs = {"class" : "_1YokD2 _3Mn1Gg"})
if target is not None:
    print("Your target area found")
else:
    print("could not find targetted area")     
```

```python
mobiles = target.find_all("div", class_ = "_2kHMtA") 
if mobiles is not None:
    print(len(mobiles), " MObiles found in page.")
else:
    print("could not find the tags you are looking for")    
```

```python
for product in mobiles:
    name = product.find("div", class_ = "_4rR01T")
    print("Name:", name.text)
    orig_price = product.find("div", class_ = "_3I9_wc _27UcVY")
    try:
        print("Original price:", orig_price.text)
    except:
        print("Original price not availale")    
    dprice = product.find("div", class_ = "_30jeq3 _1_WHN1")
    print("Discounted price:", dprice.text)
    exch_price = product.find("div", class_ = "_2Tpdn3 _18hQoS")
    try:
        print("Exchange price", exch_price.text)
    except:
        print("No exchange offer available")      

    rating = product.find("div", class_ = "_3LWZlK")
    try:
        print("Rating:", rating.text)
    except:
        print("No rating available")    

    rev_n_rat = product.find("span", class_ = "_2_R_DZ")   
    try:
        print("Reviews & Ratings", rev_n_rat.text)
    except:
        print("No Reviews & Ratings available")
    print()    
```

```python
data = [] # blank
for product in mobiles:
    name = product.find("div", class_ = "_4rR01T")
    orig_price = product.find("div", class_ = "_3I9_wc _27UcVY")
    dprice = product.find("div", class_ = "_30jeq3 _1_WHN1")
    exch_price = product.find("div", class_ = "_2Tpdn3 _18hQoS")
    rating = product.find("div", class_ = "_3LWZlK")
    rev_n_rat = product.find("span", class_ = "_2_R_DZ")    
    content = {} # empty dictionary
    content["Name"] = name.text
    content["Discount price"] = dprice.text

    try:
        content["Original price"] = orig_price.text
    except:
        content["Original price"] = dprice.text

    try:
        content["Exchange upto"] = exch_price[-1].text
    except:
        content["Exchange upto"] = None

    try:
        content["Rating"] = rating.text
    except:
        content["Rating"] = None

    try:
        content["Review count"] = rev_n_rat.text
    except:
        content["Review count"] = None       

    data.append(content)                        
```

```python
import pandas as pd
pd.DataFrame(data).to_csv("Flipkart mobiles data.csv")
```
